## Phylogenetic relationships of Hominoidea family

```r
if (!require("ape")) install.packages("ape")# this command will install ape library
#if not already found among your R packages
#activate ape library
library(ape)#library to plot and analyse phylogenetic relationships
```

```r
t10<-read.tree(text="(((((((H.sapiens),(P.paniscus,P.troglodytes)),
              (Gorilla.spp)),(Pongo.spp)),
              (Hylobates.spp,Nomascus.spp,Hoolock.spp,Symphalagus.spp))));")
plot(t10,type = "cladogram")
```

**Figure 1**. Phylogenetic relationships of the Hominoidae superfamily.

```r
plot(t10)
#add edge and node labels
edgelabels("Hominoidea 20.4Ma", 2, adj = 0)
edgelabels("Hominidae 15.7Ma", 3, adj = 0)
edgelabels("Homininae 7.3Ma", 4, adj = 0,frame = "r")
edgelabels("Hylobatidae 16.8Ma", 15, adj = 0,frame = "r")
edgelabels("Hominini 5.4Ma", 5, adj = 0,frame = "r", bg = "yellow")
nodelabels("1.7Ma", 17, 
           frame = "r", 
           bg = "yellow",
           font = 0.5,cex=1)
```

**Figure 2**. Phylogenetic relationships of the Hominoidae superfamily.This family embraces the current gorillas, chimpanzees, orangutans and humans. The common ancestor of bonobos (*Pan paniscus*) and chimpanzees (*P. troglodytes*) goes back to 1.7 Million years (Ma) and that shared by humans, and these primates 5.4Ma.
